```python
import ipywidgets as widgets
from datetime import datetime
from pathlib import Path

logname = f'IBC_Loader_log_{datetime.now().strftime("%Y-%m-%d_%H-%M-%S")}.log'
p = Path.cwd() / 'logs'
try:
    p.mkdir()
except:
    pass
log_watch = Path.cwd() / 'logs' / logname

out = widgets.Output(layout={'border': '1px solid black', 'width': '75%'})

url1 = widgets.Text(
    value=url,
    placeholder='Open connection settings or data pool (whichever applies) and paste the whole url in here',
    description='url:',
    disabled=False,
    style = {'description_width': 'initial'},
    layout=widgets.Layout(width='50%')
)
display(url1)

terms = widgets.ToggleButtons(
    options=['yes', 'no'],
    value=agreed,
    description='Do you agree to the terms?',
    disabled=False,
    button_style='info', # 'success', 'info', 'warning', 'danger' or ''
    style = {'description_width': 'initial'},
    #tooltips=['Description of slow', 'Description of regular', 'Description of fast'],
#     icons=['check'] * 3
)
display(terms)

skip = widgets.ToggleButtons(
    options=['skip', 'reload'],
    value=exclude_loaded,
    description='Do you want to skip already loaded tables?',
    disabled=False,
    button_style='info', # 'success', 'info', 'warning', 'danger' or ''
    style = {'description_width': 'initial'},
    #tooltips=['Description of slow', 'Description of regular', 'Description of fast'],
#     icons=['check'] * 3
)
display(skip)

all_string = widgets.ToggleButtons(
    options=['all string', 'auto'],
    value=as_string,
    description='column data types for non ABAP tables:',
    disabled=False,
    button_style='info', # 'success', 'info', 'warning', 'danger' or ''
    style = {'description_width': 'initial'},
    #tooltips=['Description of slow', 'Description of regular', 'Description of fast'],
#     icons=['check'] * 3
)
display(all_string)

delta_load = widgets.ToggleButtons(
    options=['full', 'delta'],
    value=delta,
    description='Load type:',
    disabled=False,
    button_style='info', # 'success', 'info', 'warning', 'danger' or ''
    style = {'description_width': 'initial'},
    #tooltips=['Description of slow', 'Description of regular', 'Description of fast'],
#     icons=['check'] * 3
)
display(delta_load)

btn=widgets.Button(description = 'run',
                   style = {'description_width': 'initial'},
                  )
display(btn)

def run(b):
    with open('cloud_upload_config.py', 'wb') as file_out:
        file_out.write(f"""# Example: url of cloud team: https://demo.eu-1.celonis.cloud/
# tenant = 'demo'
# cluster = 'eu-1'
# apikey: create key in the cloud team under "My Account"
# pool: Pool Analytics -> select target pool -> in url copy string after ui/

url = '{url1.value}'
delta = '{delta_load.value}'
as_string = '{all_string.value}'
agreed = '{terms.value}'
exclude_loaded = '{skip.value}'
logname = '{log_watch}'\n""".encode())
    with out:
        out.clear_output()
        print(f"""url = '{url1.value}'
delta = {delta_load.value}
as_string = {all_string.value}
agreed = {terms.value}
exclude_loaded = {skip.value}\n""")
        if terms.value is False:
            out.clear_output()
            print('you need to read and agree to the terms in disclaimer.md.')
            out.clear_output(wait=True)
        else:
            print('setting up schedule and executing it.')
            print(f'''run
watch tail {log_watch}
to see the output of the logfile live''')
            %run ./setup.ipynb
            out.clear_output(wait=True)
btn.on_click(run)
out
```
